```python
data = pd.get_dummies(data,columns=['sex'],drop_first=True)
```

```python
from sklearn.preprocessing import RobustScaler
robust_scaler =RobustScaler()
data[['age', 'cp','trestbps','chol','fbs','restecg','thalach','exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']]=robust_scaler.fit_transform(data[['age','cp','trestbps','chol','fbs','restecg','thalach','exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']])
```

```python
from sklearn.model_selection import train_test_split
X = data[['age', 'cp', 'trestbps', 'chol','fbs', 'restecg', 'thalach', 'exang','oldpeak', 'slope', 'ca','thal']]
y = data['target']
x_train,x_test,y_train,y_test =train_test_split(X,y,test_size=0.4,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logmodel = LogisticRegression() #instantiate LogisticRegression
logmodel.fit(x_train,y_train)
logpred =logmodel.predict(x_test)
```

```python
test_score = accuracy_score(logpred,y_test) #check the accuracy score
test_score
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(random_state =50)
dtree.fit(x_train, y_train)
train_score = dtree.score(x_train, y_train)
d_pred = dtree.predict(x_test)
from sklearn.metrics import accuracy_score,recall_score
test_score = accuracy_score(d_pred, y_test)
train_score, test_score
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,d_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators = 300)
ada.fit(x_train, y_train)
a_pred = ada.predict(x_test)
from sklearn.metrics import accuracy_score
test_score = accuracy_score(a_pred, y_test)
test_score
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 700)
rf.fit(x_train, y_train)
r_pred = rf.predict(x_test)
from sklearn.metrics import accuracy_score
test_score = accuracy_score(r_pred, y_test)
test_score
```

```python
#Tuning the value for n_estimators in the random forest model to find the most optimal one 
for i in [100,150,200,250,400,700]: #Using a for loop to iterate through the list of n_estimators
    rf = RandomForestClassifier(random_state = 0, n_estimators = i)
    rf.fit(x_train, y_train)
    train_score = rf.score(x_train, y_train)
    r_pred = rf.predict(x_test)
    from sklearn.metrics import accuracy_score
    test_score = accuracy_score(r_pred, y_test)
    print(train_score, test_score)
```

```python
from sklearn.model_selection import KFold #for K-fold cross validation
from sklearn.model_selection import cross_val_score #score evaluation
from sklearn.model_selection import cross_val_predict #prediction
kfold = KFold(n_splits=10, random_state=22) # k=10, split the data into 10 equal parts
cvmean =[]
accuracy=[]
std=[]
classifiers=['Random Forest','Decision Tree','Logistic Regression','AdaBoostClassifier']
models=[RandomForestClassifier(),DecisionTreeClassifier(),LogisticRegression(),AdaBoostClassifier()]
for i in models:
    model = i
    cv_result = cross_val_score(model,data[['age', 'cp', 'trestbps', 'chol','fbs', 'restecg', 'thalach', 'exang','oldpeak', 'slope', 'ca','thal']],data['target'] ,cv = kfold,scoring = "accuracy")
    cv_result=cv_result
    cvmean.append(cv_result.mean())
    std.append(cv_result.std())
    accuracy.append(cv_result)
new_models_dataframe2=pd.DataFrame({'CV Mean':cvmean,'Std':std},index=classifiers)       
new_models_dataframe2
```
